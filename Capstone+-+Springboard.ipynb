```python
#import WDI and language data csv
wdid = pd.read_csv('data/WDIData.csv')

```

```python
#drop rows filled with NaN values
wdid = wdid.dropna(axis=0, thresh=56)
del wdid['Unnamed: 61']
del wdid['Indicator Code']
wdid.tail()
```

```python
#explore development indicators 
wdid['Indicator Name'].value_counts(dropna=False)
```

```python
lang = pd.read_excel('data/end_lang.xlsx')
```

```python
lang.tail()
```

```python
del lang['Name in French']
del lang['Name in Spanish']
del lang['ID']
del lang['Sources']
```

```python
lang = lang.rename(columns = {'Country codes alpha 3':'country code'})
```

```python
lang.head()
```

```python
df = pd.merge(left=wdid, right=lang, left_on='Country Code', right_on='country code')
```

```python
del df['country code']
del df['Name in the language']
del df['Alternate names']
del df['ISO639-3 codes']
```

```python
df.info()
```

```python
df.set_index('Country Code', inplace=True, drop=False)
df.head(50)
```

```python
df.drop_duplicates(subset=['Country Code', 'Indicator Name'], keep='first', inplace=True)
```

```python
df.drop(df[df['Indicator Name'].str.contains("LCU")])    
```
